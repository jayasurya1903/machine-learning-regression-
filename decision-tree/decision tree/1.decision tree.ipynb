```python
import pandas as pd # it's a library used process the tubular data
```

```python
dataset= pd.read_csv("50_Startups.csv")  # function of pandas used to load the data
dataset
```

```python
dataset=pd.get_dummies(dataset) #to create dummies value for categorical text
dataset
```

```python
dataset=dataset.replace({True:1,False:0}) #replce the 0 and 1 for true,false
dataset
```

```python
dataset = dataset.drop(columns=['State_California']) # drop first column from the dataset
dataset
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]  # split input and output to get input from dataset 
independent
```

```python
dependent=dataset[['Profit']] # get output from dataset
dependent
```

```python
from sklearn.model_selection import train_test_split  # sklearn library for machine learning
x_train,x_test,y_train,y_test = train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor  # import decision tree regression for create model
regressor=DecisionTreeRegressor(criterion="absolute_error",splitter='best')
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt # used to visualization of data
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)  #input test to the model to predict output
```

```python
from sklearn.metrics import r2_score # measure the accuracy of the model
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle   # store the model
filename = "finalized_model_decision_tree.sav" 
pickle.dump(regressor,open(filename,"wb"))
```
