```python
import pandas as pd # import library 
import numpy as np
import xgboost as xg
```

```python
dataset= pd.read_csv("insurance.csv")  # function of pandas used to load the data
dataset
```

```python
dataset=pd.get_dummies(dataset) #to create dummies value for categorical text
dataset
```

```python
dataset=dataset.replace({True:1,False:0}) #replce the 0 and 1 for true,false
dataset
```

```python
dataset = dataset.drop(columns=["sex_female", "smoker_no"])# drop columns from the dataset
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]  # split input and output to get input from dataset 
independent
```

```python
dependent=dataset[['charges']] # get output from dataset
dependent
```

```python
from sklearn.model_selection import train_test_split  # sklearn library for machine learning
x_train,x_test,y_train,y_test = train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from xgboost import XGBRegressor # import xgboost regressor for create model
regressor = XGBRegressor(random_state=10, learning_rate=0.1)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)  #input test to the model to predict output
```

```python
from sklearn.metrics import r2_score # measure the accuracy of the model
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle   # store the model
filename = "finalized_model_XG_BOOST.sav" 
pickle.dump(regressor,open(filename,"wb"))
```

```python
load_model=pickle.load(open("finalized_model_XG_BOOST.sav" ,"rb")) # load the model from the file
result=load_model.predict([[50,30.970,3,1,0]])
```
